## Diccionario de datos

Hay cinco tablas en el conjunto de datos, y tendrás que usarlas todas para hacer el preprocesamiento de datos y el análisis exploratorio de datos. A continuación se muestra un diccionario de datos que enumera las columnas de cada tabla y describe los datos que contienen.

- `instacart_orders.csv`: cada fila corresponde a un pedido en la aplicación Instacart.
    - `'order_id'`: número de ID que identifica de manera única cada pedido.
    - `'user_id'`: número de ID que identifica de manera única la cuenta de cada cliente.
    - `'order_number'`: el número de veces que este cliente ha hecho un pedido.
    - `'order_dow'`: día de la semana en que se hizo el pedido (0 si es domingo).

- `products.csv`: cada fila corresponde a un producto único que pueden comprar los clientes.
    - `'product_id'`: número ID que identifica de manera única cada producto.
    - `'product_name'`: nombre del producto.
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'department_id'`: número ID que identifica de manera única cada departamento de víveres.
- `order_products.csv`: cada fila corresponde a un artículo pedido en un pedido.
    - `'order_id'`: número de ID que identifica de manera única cada pedido.
    - `'product_id'`: número ID que identifica de manera única cada producto.
    - `'add_to_cart_order'`: el orden secuencial en el que se añadió cada artículo en el carrito.
    - `'reordered'`: 0 si el cliente nunca ha pedido este producto antes, 1 si lo ha pedido.
- `aisles.csv`
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'aisle'`: nombre del pasillo.
- `departments.csv`
    - `'department_id'`: número ID que identifica de manera única cada departamento de víveres.
    - `'department'`: nombre del departamento.

```python
# ¿Todos los nombres de productos ausentes están relacionados con el departamento con ID 21?
import pandas as pd

df_product = pd.read_csv('/datasets/products.csv', sep=';')

productos_ausentes_nombre = df_product[df_product['product_name'].isna()]
try:
    df_aisles = pd.read_csv('/datasets/aisles.csv', sep=';')
except FileNotFoundError:
    print("Error: No se encontró el archivo 'aisles.csv'.")
    print("Por favor, asegúrate de tener el archivo de pasillos y ajusta la ruta si es necesario.")
    df_aisles = None # Asignar None para manejar el caso de error
try:
    df_departments = pd.read_csv('/datasets/departments.csv', sep=';') # Asume este nombre y separador
except FileNotFoundError:
    print("Error: No se encontró el archivo 'departments.csv'.")
    print("Por favor, asegúrate de tener el archivo de departamentos y ajusta la ruta si es necesario.")
    df_departments = None

if df_aisles is not None and df_departments is not None:
    productos_ausentes_con_pasillo = pd.merge(
        productos_ausentes_nombre,
        df_aisles,
        on='aisle_id',
        how='left'
    )
    productos_ausentes_con_departamento = pd.merge(
        productos_ausentes_con_pasillo,
        df_departments,
        on='department_id',
        how='left'
    )

    print("\nDepartamentos relacionados con productos con nombres ausentes:")
    print(productos_ausentes_con_departamento['department_id'].value_counts())
    if not productos_ausentes_con_departamento.empty:
        count_department_21 = productos_ausentes_con_departamento[
            productos_ausentes_con_departamento['department_id'] == 21
        ].shape[0]

        total_ausentes = productos_ausentes_con_departamento.shape[0]

        print(f"\nTotal de productos con nombre ausente: {total_ausentes}")
        print(f"Productos con nombre ausente en el departamento con ID 21: {count_department_21}")

        if total_ausentes > 0 and count_department_21 == total_ausentes:
            print("\n**Sí, todos los nombres de productos ausentes están relacionados con el departamento con ID 21.**")
        elif total_ausentes == 0:
            print("\n**No hay nombres de productos ausentes para verificar.**")
        else:
            print("\n**No, no todos los nombres de productos ausentes están relacionados con el departamento con ID 21.**")
            print("Aquí están los IDs de departamento y la cantidad de productos ausentes en cada uno:")
            print(productos_ausentes_con_departamento['department_id'].value_counts().sort_index())
    else:
        print("\nNo se encontraron productos con nombres ausentes.")
else:
    print("\nNo se pudo realizar la verificación de departamentos debido a la falta de archivos de datos (aisles.csv o departments.csv).")
```
